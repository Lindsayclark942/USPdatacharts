```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.patches import Shadow
%matplotlib inline
import os
```

## Pie Chart for USP Poverty Levels

```python
USPpovlev = 'High Poverty Area', 'Low Poverty Area'
USPpovperc = [74.5, 25.2]
#expl1 = (0.06, 0)
#color = ['red', 'blue']
```

```python
the_grid = GridSpec(1,1)
plt.subplot(the_grid[0, 0], aspect=1)
plt.pie(USPpovperc, labels=USPpovlev,autopct='%1.1f%%', radius=1) 
# explode=expl1, labeldistance=0.3, pctdistance=0.8, colors=color
#plt.legend(loc=3)
plt.text(-0.1, -1.2, 'a) USP', fontsize=12)

plt.savefig('USPpovPie.png')
#plt.show()
```

## Combined Poverty Chart

```python
USPpovlev = 'High Poverty Area', 'Low Poverty Area'
USPpovperc = [74.5, 25.2]

the_grid = GridSpec(1,3) #Creates grid

plt.subplot(the_grid[0, 0], aspect=1) #puts chart in position on grid
plt.pie(USPpovperc, labels=USPpovlev,autopct='%1.1f%%', radius=2) 
plt.text(-0.2, -2.5, 'a) USP', fontsize=12)

CSDApovlev = 'High Poverty Area', 'Low Poverty Area'
CSDApovperc = [72.6, 27.3]

plt.subplot(the_grid[0, 2], aspect=1)
plt.pie(CSDApovperc, labels=CSDApovlev,autopct='%1.1f%%', radius=2)
plt.text(0, -2.5, 'b) CSDA', fontsize=12)

plt.savefig('combinedpoverty.png')
#plt.show()
```
